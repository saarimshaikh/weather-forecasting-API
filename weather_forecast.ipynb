```python
from tkinter import *
from tkinter import ttk
from PIL import ImageTk, Image
import requests
import time
import datetime
from time import strftime

def getWeather(master):
    city = city_entry.get()
    api = "https://api.openweathermap.org/data/2.5/weather?q="+city+"&appid=06c921750b9a82d8f5d1294e1586276f"
    
    json_data = requests.get(api).json()
    condition = json_data['weather'][0]['main']
    temp = int(json_data['main']['temp'] - 273.15)
    min_temp = int(json_data['main']['temp_min'] - 273.15)
    max_temp = int(json_data['main']['temp_max'] - 273.15)
    pressure = json_data['main']['pressure']
    humidity = json_data['main']['humidity']
    wind = json_data['wind']['speed']
    sunrise = time.strftime('%I:%M:%S', time.gmtime(json_data['sys']['sunrise'] - 21600))
    sunset = time.strftime('%I:%M:%S', time.gmtime(json_data['sys']['sunset'] - 21600))
    global w
    w = condition
    label()
    final_info = condition + "\n" + str(temp) + "°C" 
    final_data = "\n"+ "\rMinimum Temperature: " + str(min_temp) + "°C" + "\n" + "Maximim Temperature: " + str(max_temp) + "°C" +"\n" + "Atmospheric Pressure: " + str(pressure) + "\n" +"Humidity: " + str(humidity) + "\n" +"Wind Speed: " + str(wind) + "\n" + "Sunrise Timing: " + sunrise + "\n" + "Sunset Timing: " + sunset
    
    my_canvas.create_text(150,370, text=final_info, font=f)
    my_canvas.create_text(200,500,text=final_data, font=f)
    
    
    api2= "https://api.openweathermap.org/data/2.5/forecast/daily?q="+city+"&cnt={cnt}&appid=06c921750b9a82d8f5d1294e1586276f"
    url = f'https://api.openweathermap.org/data/2.5/weather?q='+city+'&appid=06c921750b9a82d8f5d1294e1586276f'
    
    req = requests.get(url)
    data = req.json()
    
    name = data['name']
    lon = data['coord']['lon']
    lat = data['coord']['lat']
    
    exclude = "minute,hourly"
    
    url2 = f'https://api.openweathermap.org/data/2.5/onecall?lat={lat}&lon={lon}&exclude={exclude}&appid=06c921750b9a82d8f5d1294e1586276f'
    
    req2 = requests.get(url2)
    data2 = req2.json()
    
    days = []
    nights = []
    descr = []
    
    for i in data2['daily']:
        
        days.append(round(i['temp']['day'] - 273.15,2))
        nights.append(round(i['temp']['night'] - 273.15,2))
        descr.append(i['weather'][0]['main']+ ": "+i['weather'][0]['description'])
    
    string = f'[{city} - 8 days forecast]\n'
    
    for i in range(8):
        
        if i == 0:
            string += f'\nDay: {i+1} (Today)\n'
        elif i == 1:
            string += f'\nDay: {i+1} (tomorrow)\n'
        else:
            string += f'\nDay: {i+1}\n'
            
        string += 'Morning:' + str(days[i]) + '°C' + "\n"
        string += 'Night:' + str(nights[i]) + '°C' + "\n"
        string += 'Conditions:' + descr[i] + '\n'
    
    my_canvas.create_text(200,1100,text=string, font=("times new roman", 13, "bold"))
    
    
def label():
    if(w=="Clouds"):
        my_canvas.create_image(200,320, image=b, anchor="nw")
        
    elif(w=="Clear"):
        my_canvas.create_image(200,320, image=ag, anchor="nw")
    
    elif(w=="Rain"):
        my_canvas.create_image(200,320, image=c, anchor="nw")
        
    elif(w=="Smoke"):
        my_canvas.create_image(200,320, image=d, anchor="nw")
        
    else:#(w=="Haze"):
        my_canvas.create_image(200,320, image=e, anchor="nw")
    

def clock_date():
    display_date = strftime( "%d-%B\n%Y\n\n\n%A")
    Date.config(text=display_date)
    Date.after(1000,clock_date)
    
def clock_Time():
    display_Time = strftime("%H:%M:%S %p")
    global a
    a= strftime("%H")
    Time.config(text=display_Time)
    Time.after(1000, clock_Time)

def gretting():
    b=int(a)
   
    if (b>0 and b<3):
        app_welcome = Label(master, justify='center', text="Happy Dawn", font= ("times new roman", 29, "bold"))
        my_canvas.create_window((85,1), window=app_welcome,anchor="nw")

    elif(b>3 and b<12):
        app_welcome = Label(master, justify='center', text="Good morning", font= ("times new roman", 29, "bold"))
        my_canvas.create_window((70,1), window=app_welcome,anchor="nw")

    elif(b==12):
        app_welcome = Label(master, justify='center', text="Good After-noon", font= ("times new roman", 29, "bold"))
        my_canvas.create_window((50,1), window=app_welcome,anchor="nw")
    
    elif(b>12 and b<=15):
        app_welcome = Label(master, justify='center', text="Good After-noon", font= ("times new roman", 29, "bold"))
        my_canvas.create_window((50,1), window=app_welcome,anchor="nw")
    
    elif(b>15 and b<20):
        app_welcome = Label(master, justify='center', text="Good Evening", font= ("times new roman", 29, "bold"))
        my_canvas.create_window((72,1), window=app_welcome,anchor="nw")
    
    else:
        app_welcome = Label(master, justify='center', text="Happy Dawn", font= ("times new roman", 29, "bold"))
        my_canvas.create_window((85,1), window=app_welcome,anchor="nw")
    
    app_welcome = Label(master, justify='center', text="Thank You", font= ("times new roman", 29, "bold"))
    my_canvas.create_window((95,1490), window=app_welcome,anchor="nw")
    

def entry_clear(e):
    city_entry.delete(0, END)

master = Tk()
master.geometry("400x1550")
master.title("Weather Forecasting App")


f = ("times new roman", 15, "bold")
t = ("aerial", 20, "bold")
 

g= ImageTk.PhotoImage(file="test3img.jpg")

ab= ImageTk.PhotoImage(file="200.png")
ac= ImageTk.PhotoImage(file="logo_org1.jpg")
ad= ImageTk.PhotoImage(file="logo_org1.jpg")
ag= ImageTk.PhotoImage(file="sunny_100.jpg")
b= ImageTk.PhotoImage(file="clouds_100.jpg")
c= ImageTk.PhotoImage(file="rain_100.jpg")
d= ImageTk.PhotoImage(file="smoke_100.jpg")
e= ImageTk.PhotoImage(file="haze_100.jpg")


my_canvas = Canvas(master, width=400, height=1550, highlightthickness=0)
my_canvas.pack(side=LEFT, fill="both", expand=1)

Date = Label(master, font=("copperplate gothic",11, "bold"),background="#aaaaaa", foreground="#028A0F")
my_canvas.create_window(140,124, window=Date, anchor="nw")

Time = Label(master, font=('ds-digital',13,"bold"), background="#aaaaaa", foreground = "BLACK")
my_canvas.create_window(138,167, window=Time, anchor="nw")

clock_date()
clock_Time()
gretting()


city_entry = Entry(master, justify='center', fg="#E3242B",bg="white", border=0,  width = 21, font = ("times new roman", 17, "bold"), bd=0)
city_entry.insert(0, "Click & Enter Location.")
city_window =my_canvas.create_window(65, 270, anchor="nw", window=city_entry)


my_scrollbar = ttk.Scrollbar(my_canvas, orient= VERTICAL, command=my_canvas.yview)
my_scrollbar.pack(side=RIGHT, fill=Y, anchor="nw")
my_canvas.configure(yscrollcommand=my_scrollbar.set)
my_canvas.bind('<Configure>', lambda e: my_canvas.configure(scrollregion= my_canvas.bbox("all")))


my_canvas.create_image(0,0, image=g, anchor="nw")
my_canvas.create_image(95,55, image=ab, anchor="nw")
my_canvas.create_image(0,1, image=ac, anchor="nw")
my_canvas.create_image(334,1, image=ad, anchor="nw")
my_canvas.create_image(0,1490, image=ac, anchor="nw")
my_canvas.create_image(334,1490, image=ac, anchor="nw")

my_canvas.create_line( 0, 1, 400,   1,width=1, fill="black")
my_canvas.create_line( 0,50, 400,  50,width=1, fill="black")
my_canvas.create_line(50, 1, 50,1550,width=2, fill="black")
my_canvas.create_line(334,1, 334,1550,width=2, fill="black")
my_canvas.create_line(0,1490,334,1490,width=2, fill="black")

city_entry.bind('<Return>', getWeather)
city_entry.bind("<Button-1>",entry_clear)

master.resizable(width=False,height=True)

master.mainloop()
```
